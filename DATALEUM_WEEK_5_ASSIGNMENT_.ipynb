```python
file_path = 'TechCorp_Employee_Data.csv'
TechCorp = pd.read_csv(file_path)
TechCorp.head(5)
```

```python
# checking for missing value
TechCorp.isnull()
```

```python
TechCorp.isnull().sum()
```

```python
TechCorp.describe()
```

```python
TechCorp['Monthly Hours'] = TechCorp['Monthly Hours'].fillna(TechCorp['Monthly Hours'].median())
TechCorp['Satisfaction Score'] = TechCorp['Satisfaction Score'].fillna(TechCorp['Satisfaction Score'].mean())
TechCorp['Projects Completed'] = TechCorp['Projects Completed'].fillna(TechCorp['Projects Completed'].median())
```

```python
TechCorp.head(2)
```

```python
TechCorp.columns.tolist()
```

```python
TechCorp1 = TechCorp.drop(['Employee ID','Gender', 'Tenure', 'Absenteeism Days', 'Promotions'], axis=1)
```

```python
TechCorp1
```

```python
TechCorp1['Overworked'] =TechCorp1['Monthly Hours'].apply(lambda overworked:'Yes'
                                                                 if overworked>220 else 'No')
TechCorp1
```

```python
# Calculate efficiency score]
TechCorp1['Efficiency Score'] = TechCorp1['Monthly Hours']/ TechCorp1['Projects Completed']
TechCorp1
```

```python
Department_jobroles = TechCorp1.groupby(['Department', 'Job Role']).size().reset_index(name='Employee-count')
Department_jobroles
```

```python

plt.figure(figsize = (10, 6))
sn.barplot(data=TechCorp1, x='Department', y='Satisfaction Score')
plt.title('Average Satisfaction by Department')
plt.xlabel('Department')
plt.ylabel('Satisfaction Score')
plt.xticks(rotation=15)
plt.show()
```

```python
plt.figure(figsize = (10, 6))
sn.barplot(data=TechCorp1, x='Job Role', y='Efficiency Score')
plt.title('Efficiency Score by Job Role')
plt.xlabel('Job Role')
plt.ylabel('Efficiency Score')
plt.xticks(rotation=15)
plt.show()
```

```python
plt.figure(figsize = (10, 6))
sn.barplot(data=TechCorp1, x='Overworked', y='Department')
plt.title('Department by Overworked')
plt.xlabel('Overworked')
plt.ylabel('Department')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sn.histplot(TechCorp1['Efficiency Score'], bins=30, kde=True)
plt.title('Employee Efficiency ')
plt.xlabel('Efficiency Score')
plt.ylabel('Frequency')
plt.show()
```
